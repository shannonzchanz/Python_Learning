=================Looping Constructs (for/while)=================

=================WHILE LOOP=================

while loop syntax
while condition is still true:
    do something

```python
x = 1
while (x<10):  # while loop condition has to be true
    print(x)
    x = x + 1
    
```

```python
glasses_of_beer = 0
drunk = False

while (not drunk):    # True (not drunk)  not True == False
    glasses_of_beer+=1
    if glasses_of_beer==8:
        drunk=True
    print("{} glasses of beer drank".format(glasses_of_beer))
```

```python
processor = "amdryzenr7"  #processor[i] = a, processor[i] = m... processor[9] = 7
i = 0
print(len(processor))
while (i<len(processor)):
    print(processor[i])
    i+=1
```

```python
def addition(a, b):
    return a + b

def multiplication(a,b):
    return a*b

print(addition(3,5))
print(multiplication(1+2,3+4))
```

```python
# Positional arguments

def position_example(pc_item, kitchen_item):
    print(pc_item + " belongs to the computer")
    print(kitchen_item + " belongs to the kitchen")
    
position_example("mouse", "kitchen board")
```

```python
## default arguments have to at the tail end

def carbonara(pax, eggs_per_pax=2):
    total_eggs = pax*eggs_per_pax
    return total_eggs

print(carbonara(4,0.5))
```

```python
## default arguments have to at the tail end

def carbonara(pax, eggs_per_pax=2, bacon_per_pax=2):
    ingredients = pax*(eggs_per_pax + bacon_per_pax)
    return ingredients

print(carbonara(3,bacon_per_pax=1))
```

```python
## flexible arguments
# def addition(a,b)
#     return a + b

def addition(*args):
    total = 0
    for i in args:
        total += i
    print(total)
    return total

addition(1,2,2,4,30,230,20)
```
